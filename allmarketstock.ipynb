```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit(allmarketstock[['Name']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=encoder.categories_),['Name']),remainder='passthrough')
```

```python
column_trans
```

```python
allmarketstock
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
x_train = allmarketstock.loc[allmarketstock['Date'] <= '2018-01-01', ['Prev Close', 'Open', 'High', 'Low', 'Last', 'VWAP','Name']]
y_train = allmarketstock.loc[allmarketstock['Date'] <= '2018-01-01', 'Close']
```

```python
x_test = allmarketstock.loc[allmarketstock['Date'] > '2018-01-01', ['Prev Close', 'Open', 'High', 'Low', 'Last', 'VWAP','Name']]
y_test = allmarketstock.loc[allmarketstock['Date'] > '2018-01-01', 'Close']
```

```python
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_lr=pd.Series(y_pred,index=y_test.index)
mse = mean_squared_error(y_test, y_lr)
np.sqrt(mse)
```
