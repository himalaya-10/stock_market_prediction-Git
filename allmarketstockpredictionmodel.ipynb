```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
import pickle
for Name in sortallmarketstock['Name'].unique():
    stock_name=sortallmarketstock[sortallmarketstock['Name'] == Name]
    x_train = stock_name.loc[stock_name['Date'] <= '2018-01-01', ['Prev Close', 'Open', 'High', 'Low', 'Last', 'VWAP']]
    y_train = stock_name.loc[stock_name['Date'] <= '2018-01-01', 'Close']
#     x_test = stock_name[ap['Date'] > '2018-01-01', ['Prev Close', 'Open', 'High', 'Low', 'Last', 'VWAP']]
#     y_test = stock_name['Date'] > '2018-01-01', 'Close']
    lr.fit(x_train,y_train)
    filename = f'{Name}.pkl'
    pickle.dump(lr, open(filename, 'wb'))
    
```
